# **Import necessary library**

```python
import pandas as pd
import numpy as np
```

# **Numpy**

# **Create a NumPy array with 20 random integers between 1 and 100 and Display the NumPy array**

```python
arr = np.random.randint(1, 101, 20)
print(arr)
```

## **1. Find the sum of all elements**

```python
sum_of_elements = np.sum(arr)
print(sum_of_elements)
```

## 2. **Find the maximum and minimum values in the array**

```python
min_value = np.min(arr)
max_value = np.max(arr)
print(min_value)
print(max_value)
```

# **Pandas**

# **Create a DataFrame with 5 rows and 3 columns of random numbers**

```python
df = pd.DataFrame(np.random.randint(1, 1001, size=(5, 3)), columns=['Age', 'Salary', 'Year_of_Experience'])
print(df)
```

## **1. Compute the sum of each column**

```python
sum_of_columns = df.sum()
print(sum_of_columns)
```

## **2. Compute the mean of each column**

```python
mean_of_columns = df.mean()
print(mean_of_columns)
```
