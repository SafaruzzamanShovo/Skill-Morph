#**3. Calculate the area and perimeter of a rectangle with length 7 and width 5.**

```python
length = 7
width = 5

area = length * width
perimeter = 2 * (length + width)

print("Area:", area)
print("Perimeter:", perimeter)
```

# **9. Write a Python function that takes a string as input and prints the first and last character of the string.**

```python
def print_first_last_Char(text):
    if len(text) > 0:
        print("First character:", text[0])
        print("Last character:", text[-1])
    else:
        print("Empty string!")

print_first_last_Char("Bangladesh")
```
